```python
#FOr proper visualization, some of these columns need to be scaled.
#Though the values have relation, the difference between the value size can make it difficult to visualize the data.

#Formula to scale values: x - x.min/x.max-x.min
sdata = data.copy()
data.MPG.value_counts(normalize=True,bins=4)
#Something like this maybe
```

```python
def scale_vals(x):
    res = (x-x.min())/(x.max()-x.min())
    return res

sdata.MPG = scale_vals(sdata.MPG)
sdata.Displacement = scale_vals(sdata.Displacement)
sdata.Horsepower = scale_vals(sdata.Horsepower)
sdata.Weight = scale_vals(sdata.Weight)
sdata.Acceleration = scale_vals(sdata.Acceleration)



sdata.head()
#All scaled
```

```python
#Checking and extracting unqiue manufacturer names separately
make_list = pd.DataFrame()
make_list = sdata.index.str.extract('(^.*?)\s')
make_list[0].unique()

#As can be seen, some manufacturers are still repeating due to different spelling.
```

```python
#Now adding a new column in the dataframe
makes = sdata.index.str.extract('(^.*?)\s')
print(type(makes))
makes.columns=['Manufacturer'] #Adding a header to the Pandas dataframe

data['Manufacturer'] = makes.values #Assigning values to the Manufacture column of our Data 



#Replacing spelling variants with consistent values
data['Manufacturer'] = data['Manufacturer'].replace('toyouta','toyota')
data['Manufacturer'] = data['Manufacturer'].replace('mercedes','mercedes-benz')
data['Manufacturer'] = data['Manufacturer'].replace(['vokswagen','vw'],'volkswagen')
data['Manufacturer'] = data['Manufacturer'].replace('datsun','nissan')
data['Manufacturer'] = data['Manufacturer'].replace(['chevroelt','chevy'],'chevrolet')


data.head(10)
```
